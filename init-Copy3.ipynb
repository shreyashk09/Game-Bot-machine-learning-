```python
import numpy as np
import pandas as pd
import time
import cv2
y=200
lim = 5
lmd = 5
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
kernel = np.ones((3,3),np.uint8)
vrec = cv2.VideoCapture('v4.mp4',0)
kkk = 300
while(kkk):
    kkk-=1
    ret, scr = vrec.read()
while True:
#     scr = cv2.imread('18.png')

    ret, scr = vrec.read()
    scr = cv2.resize(scr, (680,480)) 
    shape = scr.shape
    newscr = np.zeros((shape[0],shape[1]),np.uint8)
    mx = shape[0]
    my = shape[1]//2
    # print(shape,mx,my)
    scr1 = scr 
    cv2.bilateralFilter(scr,70,30,100)
    scr = cv2.cvtColor(scr, cv2.COLOR_BGR2GRAY)
    # scr = clahe.apply(scr)
    mean = np.mean(scr[440:480,200:480]) 
    mean = np.uint8(mean)
    hor_rd(mean)

    newscr = cv2.morphologyEx(newscr,cv2.MORPH_OPEN,kernel, iterations = 1)
    newscr = cv2.morphologyEx(newscr,cv2.MORPH_CLOSE,kernel, iterations = 1)
    # newscr = cv2.morphologyEx(newscr,cv2.MORPH_OPEN,kernel, iterations = 1)
    # newscr = cv2.erode(newscr,kernel,iterations=1)
#     cv2.destroyAllWindows()
    cv2.imshow('scr',scr)
    # cv2.imshow('newscr',newscr)
    band =  cv2.bitwise_and(scr,scr,mask = newscr)

    cv2.imshow('blur',band)


    sure_bg = cv2.dilate(newscr,kernel,iterations=1)
            # Finding sure foreground area
    dist_transform = cv2.distanceTransform(newscr,cv2.DIST_L2,5)
    ret, sure_fg = cv2.threshold(dist_transform,0.0001*dist_transform.max(),255,0)
            # Finding unknown region
    sure_fg = np.uint8(sure_fg)
    unknown = cv2.subtract(sure_bg,sure_fg)
            # Marker labelling
    ret, markers = cv2.connectedComponents(sure_fg)
#     print(markers)
            # Add one to all labels so that sure background is not 0, but 1
    markers = markers+1
            # Now, mark the region of unknown with zero
    markers[unknown==255] = 0


    markers = cv2.watershed(scr,markers)
    scr[markers == -1] = [0,0,255]
    # print(markers)

    cv2.imshow('sure_fg',sure_fg)
    cv2.imshow('scr',scr)
    # cv2.imshow('sure_bg',sure_bg)
    


    q=cv2.waitKey(1)
    if q == 27:
        cv2.destroyAllWindows()
```
